```python
from datetime import datetime, timedelta

import pandas as pd
import numpy as np

import pandas_datareader as pdr


import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
from xgboost import XGBRegressor
```

## Feature Creation

```python
def get_data(ticker):
    data = pdr.get_data_yahoo(ticker, start="2000-01-01").iloc[:-1][['Close']]
    data[f"{ticker}_return"] = np.log(data.Close) - np.log(data.Close).shift(1)
    data = data.dropna()
    data = data[[f"{ticker}_return"]]
    return data
```

```python
ihsg = get_data('^JKSE')
dow_jones = get_data('^DJI')
SnP500 = get_data('^GSPC')
```

```python
df = ihsg.join(dow_jones).join(SnP500)
```

```python
df['month'] = df.index.month
df['day'] = df.index.day
df['salary_period'] = [1 if (date>=25 or date<=5) else 0  for date in df.day]
```

```python
df = df.dropna()
```

## Split data

```python
train = df.loc[df.index < '2018-01-01']
test = df.loc[df.index >= '2018-01-01']
```

```python
X_train = train.drop(columns="^JKSE_return")
y_train = train[['^JKSE_return']]
```

```python
X_test = test.drop(columns='^JKSE_return')
y_test = test[['^JKSE_return']]
```

```python
fig, ax = plt.subplots(figsize=(15,5))

ax.plot(y_train, label="train_data", color="blue")
ax.plot(y_test, label="test_data", color="red")
ax.legend()
ax.set_title("IHSG Return Train and Test Data")

fig.savefig('pic/ihsg_return_train_test.png')
```

## Modelling

```python
reg = XGBRegressor(n_estimators=500,
                   early_stopping_rounds=50)

reg.fit(X_train, y_train, 
        eval_set=[(X_train, y_train), (X_test, y_test)], 
        verbose=True)
```

## Feature Importance

```python
fi = pd.DataFrame(data=reg.feature_importances_, 
              index=reg.feature_names_in_, 
              columns=['importance'])
```

```python
fi.sort_values('importance').plot(kind='barh', 
                                  title='Feature Importance', 
                                  figsize=(10, 5))

plt.savefig('pic/feature_importance.png')
```

## Forecast

```python
test['prediction'] = reg.predict(X_test)
```

```python
df = df.merge(test[['prediction']], how="left", left_index=True, right_index=True)
```

```python
ax = df[['^JKSE_return']].plot(figsize=(20, 10))
df['prediction'].plot(ax=ax, label="preds")

plt.legend()
plt.title("IHSG Return Prediction")

plt.savefig('pic/prediction.png')
```

```python
test[['^JKSE_return']].plot(figsize=(20,10))
df['prediction'].plot(label="preds")

plt.legend()
plt.title("IHSG Return Prediction Zoom in")
```
